```python
# remove warnings
import warnings
warnings.filterwarnings('ignore')
# ---

%matplotlib inline
import pandas as pd
pd.options.display.max_columns = 100
from matplotlib import pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import numpy as np

pd.options.display.max_rows = 100

# load the posting data
data = pd.read_csv('Total-scrape.csv',delimiter='#')

# remove duplicate job descriptions
print('Before duplicate removal:',data.shape)
data = data.drop_duplicates(subset='Description')
print('After duplicate removal:',data.shape)

data.head()
```

```python
from wordcloud import WordCloud

# We'll remove these words from the comment strings before making wordclouds.
removewords = (['Job ','Description ','Title ','Data Scientist','Data Science','data scientist',
                'data science', 'sexual orientation', 'gender identity', 'years experience', 
                'qualified applicants', 'national origin','race color', 'veteran status',
               'without regard', 'days ago', 'days', 'day', 'day ago', 'day', 'Portland', 'Beaverton','Jobs', 'Job', 
                'equal opportunity','Glassdoor'])

# concat all text into a single string
wordstr = data.Description.str.cat()

# remove common words
for word in removewords:
    wordstr = wordstr.replace(word, '')

# Generate a word cloud image
wordcloud = WordCloud(width=400, height=400,colormap='plasma').generate(wordstr)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Job description word cloud\n", fontsize=28)
plt.axis("off");
```

```python
# concat all text into a single string
wordstr = data.Position.str.cat()

# remove common words
for word in removewords:
    wordstr = wordstr.replace(word, '')

# Generate a word cloud image
wordcloud = WordCloud(width=400, height=400,colormap='plasma').generate(wordstr)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Position name word cloud\n", fontsize=28)
plt.axis("off");
```
